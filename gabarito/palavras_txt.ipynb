```python
import random

# Abre o arquivo em modo de leitura ('r' = read)
with open('palavras.txt', 'r') as arquivo:
    # Lê todas as linhas e guarda em uma lista
    lista_de_palavras = arquivo.readlines()

# Sorteia uma palavra da lista
palavra_sorteada = random.choice(lista_de_palavras)

# Remove espaços vazios ou quebras de linha invisíveis (o comando .strip() faz isso)
palavra_secreta = palavra_sorteada.strip().upper()

print("O computador sorteou uma palavra do arquivo!")
# Agora o jogo pode começar usando a variável palavra_secreta...
```
