## 🛠️ Aquecimento: Nosso primeiro comando
Para o computador falar com a gente, usamos o comando `print`.
O código abaixo é um bloco executável. Clique no botão de "Play" (a setinha ao lado esquerdo do código) para testar!

```python
print("Olá, Mundo! Estamos prontos para programar o nosso jogo.")
```

## 🧩 Etapa 1: Variáveis e Perguntas
Como fazemos o computador guardar o nosso chute? Usamos **variáveis**!
Execute o bloco abaixo, digite o seu número na caixinha que vai aparecer e aperte a tecla "Enter".


```python
numero_secreto = 7
chute = int(input("Qual é o seu chute (entre 1 e 10)? "))
```

## ⚖️ Etapa 2: Acertou ou Errou? (If / Else)
Agora precisamos que o computador verifique se o número secreto é igual ao nosso chute. Para isso, usamos uma estrutura condicional (Se / Senão).

```python
if chute == numero_secreto:
  print("🏆 Parabéns! Você acertou!")
else:
  print("❌ Que pena, você errou!")
```

## 🔄 Etapa 3: O Jogo Completo (While)
Um jogo de verdade tem mais de uma tentativa, certo? Vamos usar o comando `while` (enquanto) para criar uma estrutura de repetição e dar 3 chances ao jogador!

```python
numero_secreto = 7
tentativas = 3

while tentativas > 0:
  print("\nVocê tem", tentativas, "tentativas.")
  chute = int(input("Qual é o seu chute? "))

  if chute == numero_secreto:
    print("🏆 Parabéns! Você acertou em cheio!")
    break
  else:
    print("❌ Você errou!")

  tentativas = tentativas - 1

if tentativas == 0:
  print("\n💀 Fim de jogo! O número era", numero_secreto)
```
