## 🎒 Aquecimento: Conhecendo as Listas e o laço `for`
Uma lista em Python é criada usando colchetes `[ ]`. Para passar por cada item dessa lista, usamos um novo comando de repetição chamado `for` (para cada).
Execute o código abaixo e veja a mágica acontecer:

```python
mochila = ["Espada", "Escudo", "Poção de Vida"]

print("Itens no meu inventário:")
for item in mochila:
  print("-", item)
```

## 🧩 Etapa 1: O "Tabuleiro" do Jogo
No Jogo da Forca, nós não mostramos a palavra logo de cara. Nós mostramos tracinhos `_` no lugar das letras.
Vamos usar a função `len()` (que mede o tamanho de uma palavra) para criar o nosso tabuleiro secreto!

```python
palavra_secreta = "PYTHON"
tamanho = len(palavra_secreta)

# Multiplicamos uma lista contendo um tracinho pelo tamanho da palavra
letras_descobertas = ["_"] * tamanho

print("A palavra secreta tem", tamanho, "letras!")
print(letras_descobertas)
```

## 🔍 Etapa 2: O chute está na palavra?
Como o computador sabe se o jogador acertou uma letra? Em Python, existe um comando super poderoso chamado `in` (dentro). Ele verifica se uma letra está *dentro* de uma palavra. Testem o código abaixo:

```python
palavra_secreta = "PYTHON"
chute = input("Digite uma letra: ").upper() # O .upper() transforma tudo em MAIÚSCULA

if chute in palavra_secreta:
  print("✅ Boa! Essa letra existe na palavra.")
else:
  print("❌ Ixi... Essa letra não existe.")
```

## 🚀 Etapa 3: O Jogo Completo!
Agora vamos juntar o `while` da Aula 1 com as `Listas` e o `for` da Aula 2.
Sempre que o jogador acertar, o código vai procurar em qual "posição" (índice) a letra está e trocar o `_` pela letra verdadeira.

```python
palavra_secreta = "HACKER"
letras_descobertas = ["_", "_", "_", "_", "_", "_"]
tentativas = 5

print("Bem-vindo ao Jogo da Forca!")

# O jogo roda ENQUANTO houver tentativas E houver "_" escondidos
while tentativas > 0 and "_" in letras_descobertas:
    print("\nPalavra:", " ".join(letras_descobertas)) # Junta a lista com espaços
    print("Tentativas restantes:", tentativas)

    chute = input("Digite uma letra: ").upper()

    if chute in palavra_secreta:
        print("✅ Você acertou a letra!")
        # Esse laço for passa por todas as posições da palavra
        for posicao in range(len(palavra_secreta)):
            if palavra_secreta[posicao] == chute:
                letras_descobertas[posicao] = chute
    else:
        print("❌ Letra errada!")
        tentativas -= 1

# Verificando o resultado final
if "_" not in letras_descobertas:
    print("\n🏆 PARABÉNS! Você salvou o boneco! A palavra era:", palavra_secreta)
else:
    print("\n💀 GAME OVER! Você foi enforcado. A palavra era:", palavra_secreta)
```
